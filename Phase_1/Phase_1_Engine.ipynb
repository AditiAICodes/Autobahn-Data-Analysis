```python
import time
import random
from datetime import datetime



def read_sensors():
    # The Waiter: Just gets the raw data
    speed = random.randint(120, 250)
    temp = random.randint(90, 130)
    oil_pressure=random.randint(30,80)
    return speed, temp, oil_pressure

def analyze_status(speed, temp,oil_pressure):
    # The Chef: Takes the data and runs the logic
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    if speed > 220:
        status = "\033[40m⚠️ POLICE RADAR DETECTED: SLOW DOWN\033[0m"
        log_message = f"[{now}] WARNING: Police Radar at {speed} km/h"
    elif temp > 120:
        status = "\033[91m🔴 CRITICAL OVERHEAT: INITIATING COOLING\033[0m"
        log_message = f"[{now}] CRITICAL: Engine Overheat at {temp}C"
    elif temp > 105:
        status = "\033[93m🟡 WARNING: HIGH HEAT DETECTED\033[0m"
        log_message = f"[{now}] WARNING: High Heat at {temp}C"

    elif oil_pressure > 75:
        status = "\033[91m🔴 CRITICAL: OIL PRESSURE TOO HIGH\033[0m"
        log_message = f"[{now}] WARNING: High Oil Pressure at {oil_pressure}atm"
    else:
        status = "\033[92m🟢 OPTIMAL\033[0m"
        log_message = None

    return status, log_message # Hands the final result back to the main engine

def save_to_blackbox(log_message):
    # The Manager: Saves the file if there is a warning
    if log_message:
        with open("telemetry_logs.txt", "a") as file:
            file.write(log_message + "\n")



print("\033[1m--- INITIALIZING MUNICH TELEMETRY PIPELINE V3 (MODULAR) ---\033[0m\n")
time.sleep(1)

while True:
    try:
        # 1. The Waiter gets the data
        current_speed, current_temp,current_oil_pressure = read_sensors()

        # 2. The Chef analyzes it
        current_status, message_to_save = analyze_status(current_speed, current_temp,current_oil_pressure)

        # 3. Print the dashboard
        print(f"[LIVE LOG] Speed: {current_speed} km/h | Engine Temp: {current_temp}°C |Oil Pressure:{current_oil_pressure} atm | Status: {current_status}")

        # 4. The Manager saves it (if necessary)
        save_to_blackbox(message_to_save)

        time.sleep(1.5)

    except KeyboardInterrupt:
        print("\n\033[1m--- MANUAL OVERRIDE: SYSTEM SHUTDOWN. LOGS SECURED. ---\033[0m")
        break
```

```python
import time

print("\033[1m--- INITIATING BLACK BOX FORENSIC ANALYSIS ---\033[0m\n")
time.sleep(1)

# We create some empty buckets to count our errors
police_count = 0
overheat_count = 0
pressure_count=0

try:
    # 1. Open the vault in "r" (Read) mode!
    with open("telemetry_logs.txt", "r") as file:

        # 2. Grab every single line from the file and save it
        all_logs = file.readlines()

        print(f"Total entries found in Black Box: {len(all_logs)}\n")

        # 3. Read through the lines one by one (The AI Scanner)
        for line in all_logs:

            # If the line contains the word "Police", add 1 to the bucket
            if "Police" in line:
                police_count += 1

            # If the line contains the word "CRITICAL", add 1 to the bucket
            elif "CRITICAL" in line:
                overheat_count += 1

            elif "Oil" in line:
                pressure_count += 1

    # 4. Print the final forensic report
    print("\033[94m[ FINAL FORENSIC REPORT ]\033[0m") # 94m makes it Blue!
    print(f" Police Radar Warnings: {police_count}")
    print(f"🔥 Critical Overheats:   {overheat_count}")
    print(f"🔥 Critical Oil Pressure:   {pressure_count}")

except FileNotFoundError:
    # A safety net just in case the file doesn't exist yet!
    print("\033[91mERROR: telemetry_logs.txt not found. Did you run the engine first?\033[0m")
```
